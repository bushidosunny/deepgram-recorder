```python
def format_transcript(response_json):
    # Extract the words from the JSON structure
    words = response_json['results']['channels'][0]['alternatives'][0]['words']

    # Initialize variables
    transcript_lines = []
    current_speaker = None

    # Iterate through each word in the response
    for word_info in words:
        speaker = word_info['speaker']
        punctuated_word = word_info['punctuated_word']

        # Check if speaker has changed
        if current_speaker != speaker:
            if current_speaker is not None:
                transcript_lines.append('')  # Add a blank line to separate speakers
            current_speaker = speaker
            transcript_lines.append(f'Speaker {speaker}:')

        # Append the word to the current speaker's line
        transcript_lines[-1] += ' ' + punctuated_word

    # Join all lines into a single formatted string
    formatted_transcript = '\n'.join(transcript_lines)
    return formatted_transcript
```
